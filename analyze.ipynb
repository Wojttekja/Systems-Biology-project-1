# Lambdas mean and std

```python
import pandas as pd
import seaborn as sns
```

```python
dir_one = pd.read_csv("results/non zero adaptive directional, slower env pace_20260330_200001/summary.csv")
dir_fast = pd.read_csv("results/adaptive directional, faster env pace_20260330_192847/summary.csv")
```

```python
import matplotlib.pyplot as plt

df = dir_one  # change to dir_fast if needed

plt.figure(figsize=(10, 6))
colors = sns.color_palette("tab10", 4)

for i, color in enumerate(colors):
    mean_col = f"mean_lambda_{i}_mean"
    std_col = f"mean_lambda_{i}_std"
    
    sns.lineplot(data=df, x="generation", y=mean_col, label=f"lambda {i}", color=color)
    plt.fill_between(
        df["generation"],
        df[mean_col] - df[std_col],
        df[mean_col] + df[std_col],
        color=color,
        alpha=0.2
    )

plt.title("Mean lambda per dimension over generations")
plt.xlabel("Generation")
plt.ylabel("Lambda mean")
plt.grid()
plt.legend(title="Dimension", loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

def collect_gen_values(folder, generation, column="mean_fitness"):
    folder_path = Path(folder)
    csv_paths = sorted(p for p in folder_path.glob("*.csv") if p.name != "summary.csv")
    values = []
    for csv_path in csv_paths:
        df = pd.read_csv(csv_path)
        gen_values = df.loc[df["generation"] == generation, column].values
        if len(gen_values) == 0:
            continue
        values.append(gen_values[0])
    if not values:
        raise ValueError(f"No replicate values found in {folder_path} for generation {generation}")
    return values

dir_one_folder = "results/non zero adaptive directional, slower env pace_20260330_200001"
dir_two_dim_folder = "results/non zero adaptive directional, slower env pace, two dimensions_20260330_200255"
iso_one_folder = "results/isotropic,one dim_20260330_202459"
iso_two_folder = "results/non zero adaptive directional, slower env pace, two dimensions_20260330_200255"

conditions = [
    "adaptive dir\n[0.25, 1e-7, 1e-7, 1e-7]",
    "adaptive dir\n[0.125, 0.125, 1e-7, 1e-7]",
    "isotropic\n[0.25, 1e-7, 1e-7, 1e-7]",
    "isotropic\n[0.125, 0.125, 1e-7, 1e-7]",
 ]

folders = [dir_one_folder, dir_two_dim_folder, iso_one_folder, iso_two_folder]

gen_50 = [collect_gen_values(folder, 50) for folder in folders]
gen_100 = [collect_gen_values(folder, 100) for folder in folders]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.violinplot(gen_50, positions=range(len(conditions)), showmeans=True)
ax1.set_xticks(range(len(conditions)))
ax1.set_xticklabels(conditions)
ax1.set_ylabel("Mean Fitness")
ax1.set_title("Fitness at Generation 50")
ax1.grid(axis="y", alpha=0.3)
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')

ax2.violinplot(gen_100, positions=range(len(conditions)), showmeans=True)
ax2.set_xticks(range(len(conditions)))
ax2.set_xticklabels(conditions)
ax2.set_ylabel("Mean Fitness")
ax2.set_title("Fitness at Generation 100")
ax2.grid(axis="y", alpha=0.3)
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
